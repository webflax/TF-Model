```python
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('btc.moh','rb') as f :
    X = pickle.load(f)
XO = X.close
# XO = np.log(XO)   # log
```

```python
'''
ùë•ùëõùëúùëüùëö = 0.5 [ùë°ùëéùëõ‚Ñé [0.01(ùë• ‚àí ùúá)ùõø] + 1] 
'''

std = np.std(XO)

mean = np.mean(XO)

def Tanh_estimators(x,s=std,m=mean):
    return ( 0.5 * ( ( np.tanh( ( 0.01 * ( x - m ) ) / s ) ) + 1 ) )

def Tanh_estimators_invers(x,s=std,m=mean):
    return ( ( ( (np.arctanh( x * 2 - 1 ) )  * s ) / 0.01 ) + m )
```

```python
XN = XO.map(Tanh_estimators)
```

```python
XI = XN.map(Tanh_estimators_invers)
```

```python
plt.plot(XN)
```
